```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python


# Sampling frequency
fs = 1000
# Cutoff frequency
fc = 21.5 #fmax=540*0.50/2*3.14*0.2=215,fc=fmax/10=21.5Hz
n=2 #filter order
# Calculate normalized cutoff frequency
f_c_norm = fc / (fs/2)

# Design low-pass Butterworth filter
b, a = signal.butter(n, f_c_norm, btype='low')

# Generate sample signal 1 (sinusoid at 20 Hz, with random noise)
t = np.arange(0, 2, 1/fs)
x1 = np.sin(2 * np.pi * 20 * t) + 0.3 * np.random.randn(len(t))

# Generate sample signal 2 (sinusoid at 30 Hz, with random noise)
x2 = np.sin(2 * np.pi * 30 * t) + 0.3 * np.random.randn(len(t))

# Apply filter to both signals
y1 = signal.lfilter(b, a, x1)
y2 = signal.lfilter(b, a, x2)

# Plot signal 1 (original and filtered)
fig, ax = plt.subplots(2, 1, figsize=(8, 5))
ax[0].plot(t, x1, label='Signal 1 (Original)')
ax[0].set_title('Signal 1')
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[0].legend()
ax[1].plot(t, y1, label='Signal 1 (Filtered)', color='green')
ax[1].set_title('Filtered Signal 1')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Amplitude')
ax[1].legend()

# Plot signal 2 (original and filtered)
fig, ax = plt.subplots(2, 1, figsize=(8, 5))
ax[0].plot(t, x2, label='Signal 2 (Original)')
ax[0].set_title('Signal 2')
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[0].legend()
ax[1].plot(t, y2, label='Signal 2 (Filtered)', color='green')
ax[1].set_title('Filtered Signal 2')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Amplitude')
ax[1].legend()

# Display the plot
plt.show()
```
