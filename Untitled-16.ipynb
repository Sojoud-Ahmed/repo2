```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python

# Sampling frequency
fs = 1000
# Cutoff frequency
fc = 20
n=2 #filter order
# Calculate normalized cutoff frequency
f_c_norm = fc / (fs/2)

# Design low-pass Butterworth filter
b, a = signal.butter(n, f_c_norm, btype='low')

# Generate sample signal 1 (sinusoid at 20 Hz, with random noise)
t = np.arange(0, 2, 1/fs)
x1 = np.sin(2 * np.pi * 20 * t) + 0.5 * np.random.randn(len(t))

# Generate sample signal 2 (sinusoid at 60 Hz, with random noise)
x2 = np.sin(2 * np.pi * 60 * t) + 0.5 * np.random.randn(len(t))

# Apply filter to both signals
y1 = signal.lfilter(b, a, x1)
y2 = signal.lfilter(b, a, x2)

# Plot both signals (original and filtered) in the same figure
fig, ax = plt.subplots(2, 1, figsize=(9, 5))

# Plot the original signals
ax[0].plot(t, x1, label='Signal 1 (Original)')
ax[0].plot(t, x2, label='Signal 2 (Original)')
ax[0].set_title('Original Signals')
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[0].legend()

# Plot the filtered signals
ax[1].plot(t, y1, label='Signal 1 (Filtered)')
ax[1].plot(t, y2, label='Signal 2 (Filtered)')
ax[1].set_title('Filtered Signals')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Amplitude')
ax[1].legend()

# Display the plot
plt.show()


```
